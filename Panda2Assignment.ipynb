```python
import pandas as pd
import numpy as np
```

<B>Read the dataset from the below link</B>

```python
df_Data= pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')
print(df_Data.info())
df_Data.head(2)
```

<B>1. Delete unnamed columns</B>

```python
# 1. Delete unnamed columns

df_Data_DeletdCol= df_Data.drop(['Unnamed: 0'],axis=1)

df_Data_DeletdCol.head()
```

<B>2. Show the distribution of male and female</B>

```python
#shows without converting to percentage distribution
df_Data_DeletdCol['Gender'].value_counts(normalize=True)
```

```python
#shows the count of Gender
df_Data_DeletdCol.groupby(['Gender']).size().reset_index(name='counts')
```

```python
#shows the percentage distribution
round(df_Data['Gender'].value_counts(normalize=True)*100,2)
```

```python
gender_dist=pd.Categorical(df_Data_DeletdCol["Gender"])
```

```python
gender_dist=gender_dist.rename_categories(["Female","Male"])
```

```python
print("distribution of male and female :\n")
gender_dist.describe()
```

<B>3. Show the top 5 most preferred names</B>

```python
#selects Name, COunt group by name and sorts the top 5 by accending order
df_Data_DeletdCol[["Name","Count"]].groupby('Name').sum().sort_values("Count",ascending=0).head(5)
```

```python
sum_of_count=df_Data_DeletdCol[["Name","Count"]].groupby("Name",as_index=True).sum()
```

```python
print("Showing the top 5 most preferred names")
sum_of_count.sort_values(by="Count",ascending=False)[:5]
```

<B>4. What is the median name occurence in the dataset</B>

```python
print("Median of name occurance: ",sum_of_count.median()[0])
```

```python
df_Data_median_Name=df_Data_DeletdCol[["Name","Count"]].groupby('Name').sum()
```

```python
print("Median name occurence in the dataset the first 5 records only \n")
df_Data_median_Name[df_Data_median_Name['Count'] == df_Data_median_Name['Count'].median()].head()
```
